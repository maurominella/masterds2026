# WHILE loops

```python
counter = 0
while counter < 5:
    print(f"within the loop: {counter}")
    counter  += 1
    if (counter == 3):
        break
        
print(f"outside of the loop: {counter}")
```

```python
for i in range(3): # loop for 3 rows    
    for j in range(5): # loop for 5 columns
        print("*", end="")
    print()
    
```

# Error Handling

```python
try:
    x = 3 / 0

except ZeroDivisionError as e:
    print(e)
```

```python
try:
    x = 10 + "abc"

except TypeError as e:
    print(f"The following error occurred: {e}")
```

```python
try:
    my_list = [1,2,3,4,5]
    print(my_list[10])

except Exception as e:
    print(f"The following error type has happened: {e}")
```

```python
try:
    x = 3 / 0

except ZeroDivisionError as e:
    print(e)
```

```python
try:
    x = 10
    x = 10 + "abc"

except TypeError as e:
    print(e)
```
